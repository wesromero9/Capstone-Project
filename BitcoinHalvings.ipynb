```python
import pandas as pd
import hvplot.pandas
```

```python
df = pd.read_csv("Bitcoin.csv")
df = df.set_index(pd.DatetimeIndex(df["Date"].values))
df
```

```python
#Bitcoin Historical
btc_df = df
btc_df = btc_df.drop(columns = ["Date","Change %"],axis=1)
btc_df.head()
```

```python
# Plot the price action before and after the first halving event
btc_plot = btc_df.hvplot(title="Bitcoins Historical Price",
                        grid=True,xlabel='Date',ylabel='Bitcoin Price',color="black",legend="top",fontsize=12)
btc_plot
```
