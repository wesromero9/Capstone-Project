```python
import pandas as pd
import hvplot.pandas
import numpy as np
import panel as pn
import plotly.express as px
pn.extension('plotly')
```

```python
daily_returns = pd.read_csv("BitcoinSP500.csv")
daily_returns = daily_returns.set_index("Date")
daily_returns = daily_returns.pct_change()
daily_returns = daily_returns.dropna()
daily_returns
```

```python
# Visualize distribution of Tesla percent change in closing price using a histogram plot
daily_returns["Bitcoin Price"].plot.hist()
```

```python
# Visualize distribution of SP500 percent change in closing price using a histogram plot
daily_returns["S&P500 Price"].plot.hist()
```

```python
# Visualize the distribution of daily returns across all stocks using a histogram plot
daily_returns.plot.hist(alpha=1)
```
