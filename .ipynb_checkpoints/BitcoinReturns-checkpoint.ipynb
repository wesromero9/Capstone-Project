```python
# import dependencies
import pandas as pd
import hvplot.pandas
from bokeh.models.formatters import BasicTickFormatter
import panel as pn
from panel.interact import interact
from panel import widgets
pn.extension()
```

```python
df = pd.read_csv("Bitcoin.csv")
df = df.set_index(pd.DatetimeIndex(df["Date"].values))
df_2 = df.drop(columns=["Date","Change %"],axis=1)
df_2
```

# Bitcoins Daily Returns

```python
# Calculate the daily return using the 'shift()' function
daily_returns = df_2.pct_change()
daily_returns.to_csv("BtcReturns.csv")
# Plot pct change vs Prcice
daily_returns = daily_returns.dropna()
#daily_returns.drop("Price Change",axis=1,inplace=True)
daily_returns = daily_returns.rename(columns = {'Price':'Daily Returns'})
daily_returns
```

```python
# Plot the daily returns of Bitcoin
daily_returns.hvplot(figsize=(10,5), xlabel="Date",ylabel="Daily Returns", title="Bitcoin Historical Daily Returns")
```

```python
# daily returns during bitcoin halving Nov 28 2012
halving_1 = daily_returns.loc['2011-11-28':'2014-11-28']
halving_1 = halving_1.dropna()
halving_1

# daily returns during bitcoin halving July 9 2016
halving_2 = daily_returns.loc['2015-07-09':'2018-07-09']
halving_2

# daily returns during bitcoin halving May 11 2020
halving_3 = daily_returns.loc['2019-05-11':'2022-02-04']
halving_3
```

```python
# Display chart of Bitcoin daily returns during bitcoins 1st halving event. Nov 28 2012
halving_1.hvplot(xlabel="Date",ylabel="Daily Returns (%)", title="Bitcoin Daily Returns During 1st Halving Event")
```

```python
# Display chart of Bitcoin daily returns during bitcoins 2nd halving event. July 9 2016
halving_2.hvplot(xlabel="Date",ylabel="Daily Returns (%)", title="Bitcoin Daily Returns During 2nd Halving Event")
```

```python
# Display chart of Bitcoin daily returns during bitcoins 3rd halving event. May 11 2020
halving_3.hvplot(xlabel="Date",ylabel="Daily Returns (%)", title="Bitcoin Daily Returns During 3rd Halving Event")
```

```python
# Group by year, month, and day and grab first of each group
halving_1_split = halving_1.groupby([halving_1.index.year, halving_1.index.month, halving_1.index.day]).first()
halving_1_split.plot(rot=47)
```

```python
# Group by year, month, and day and grab first of each group
halving_2_split = halving_2.groupby([halving_2.index.year, halving_2.index.month, halving_2.index.day]).first()
halving_2_split.plot(rot=47)
```

```python
# Group by year, month, and day and grab first of each group
halving_3_split = halving_3.groupby([halving_3.index.year, halving_3.index.month, halving_3.index.day]).first()
halving_3_split.plot(rot=47)
```

# Bitcoins Cumulative Returns

```python
# Calculate the cumulative returns using the 'cumprod()' function
cumulative_returns = (1 + daily_returns).cumprod()
cumulative_returns.head()
```

```python
# Plot the daily returns of the Bitcoin over the last 5 years
#cumulative_returns.hvplot(figsize=(10,5))
cumulative_plot = cumulative_returns.hvplot(title="Bitcoins Cumulative Returns",
                        grid=True,xlabel='Date',ylabel='Bitcoin Price',color="green",legend="top",fontsize=12)
cumulative_plot
```

# Price Vs Daily Returns Scatter Plot

```python
price_vs_dr = pd.merge(df, daily_returns, how='inner', left_index=True, right_index=True).reset_index()
price_vs_dr = price_vs_dr.drop(columns=["index","Date","Change %"],axis=1).dropna()
price_vs_dr = price_vs_dr.rename(columns = {'Price_x':'Bitcoin Price'})
price_vs_dr = price_vs_dr.rename(columns = {'Price_y':'Daily Returns'})
price_vs_dr = price_vs_dr.set_index("Bitcoin Price")
```

```python
# Plot Bitcoin Price vs Daily Returns
price_vs_dr_plot = price_vs_dr.hvplot(kind="scatter",title="Bitcoin Price vs Daily Returns",yformatter='%.0f')
price_vs_dr_plot
```
